# 2. Youtube Comments Scraper

Based on the Youtube videos url scraped from the previous step (1. youtube video scraper), using APIFY scrape the cooment in each videos.

```python
import pandas as pd
import os
from dotenv import load_dotenv
from apify_client import ApifyClient
import json
```

```python
df = pd.read_csv('../data/youtube_videos.csv')
df = df[df['commentsCount'] > 0]
df.head()
```

```python
df = df[['url', 'input']]
print("============================================")
print(df.head())
print("============================================")
print(df.info())
print("============================================")
print(df.input.value_counts())
```

```python
load_dotenv()

APIFY_API_KEY = os.getenv("APIFY_API_KEY")
client = ApifyClient(APIFY_API_KEY)

start_urls = [{"url": u} for u in df["url"].dropna().tolist()]

run_input = {
    "startUrls":start_urls,
    "maxComments": 300,
    "commentsSortBy": "0",
}

run = client.actor("p7UMdpQnjKmmpR21D").call(run_input=run_input)

for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    print(item)
```

```python
data = []
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    data.append(item)
with open("../data/youtube_comments.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("Saved results to ../data/youtube_comments.json")
```

```python
# https://console.apify.com/actors/p7UMdpQnjKmmpR21D/input
INPUT_JSON = "../data/youtube_comments.json"
OUTPUT_CSV = "../data/youtube_comments.csv"

with open(INPUT_JSON, "r", encoding="utf-8") as f:
    data = json.load(f)

df_comments = pd.json_normalize(data)

columns_needed = [
    "cid",
    "pageUrl",
    "comment",
    "replyCount"
]

df_comments = df_comments[[col for col in columns_needed if col in df_comments.columns]]
df_comments.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")

print(f"✅ CSV file saved as '{OUTPUT_CSV}' with {len(df_comments)} rows.")
```

```python
merged_df = df_comments.merge(
    df[['url', 'input']],        
    left_on='pageUrl',         
    right_on='url',            
    how='left'                
)

merged_df = merged_df.drop(columns=['url'])

print(merged_df.head())
print(merged_df.info())

df_cleaned = merged_df[['input','comment','replyCount']]
df_cleaned.to_csv('../data/youtube_comments_cleaned.csv', index=False, encoding="utf-8-sig")
```
