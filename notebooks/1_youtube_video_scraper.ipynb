# 1. Youtube Video Scraper

Using APIFY API scrape the youtube vidoes related with smarphones review.

```python
import pandas as pd
import json
import os
from dotenv import load_dotenv
from apify_client import ApifyClient
```

```python
# https://console.apify.com/actors/h7sDV53CddomktSi5/input

load_dotenv()

APIFY_API_KEY = os.getenv("APIFY_API_KEY")
# Initialize the ApifyClient with your API token
client = ApifyClient(APIFY_API_KEY)

# Prepare the Actor input
run_input = {
    "searchQueries": ["Iphone 17","Samsung Galaxy Z Fold 7"],
    "maxResults": 50,
    "maxResultsShorts": 0,
    "maxResultStreams": 0,
    "startUrls": [],
    "subtitlesLanguage": "en",
    "subtitlesFormat": "srt",
}

# Run the Actor and wait for it to finish
run = client.actor("h7sDV53CddomktSi5").call(run_input=run_input)

# Fetch and print Actor results from the run's dataset (if there are any)
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    print(item)
```

```python
data = []

for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    data.append(item)

with open("../data/youtube_videos.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("Saved results to ../data/youtube_videos.json")
```

```python

INPUT_JSON = "../data/youtube_videos.json"
OUTPUT_CSV = "../data/youtube_videos.csv"

with open(INPUT_JSON, "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.json_normalize(data)

columns_needed = [
    "title",
    "url",
    "viewCount",
    "date",
    "likes",
    "channelUrl",
    "channelId",
    "numberOfSubscribers",
    "commentsCount",
    "input"
]

df = df[[col for col in columns_needed if col in df.columns]]


df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")

print(f"CSV file saved as '{OUTPUT_CSV}' with {len(df)} rows.")
```
